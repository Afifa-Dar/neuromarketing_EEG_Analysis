```python
import pywt
import numpy as np
import pandas as pd

# Assuming you have EEG data stored in a Pandas DataFrame called eeg_data
# Each column represents a channel, and the sampling rate is 128 Hz

# Define the wavelet type (db4)
wavelet = 'db4'

# Define the frequency bands of interest
frequency_bands = {
    'Delta': (1, 4),    # Delta band: 0.5 Hz - 4 Hz
    'Theta': (4, 8),      # Theta band: 4 Hz - 8 Hz
    'Alpha': (8, 13),     # Alpha band: 8 Hz - 13 Hz
    'Beta': (13, 22),     # Beta band: 13 Hz - 30 Hz
    'Gamma': (32, 100)    # Gamma band: 30 Hz - 100 Hz
}

# Convert the EEG data DataFrame to a NumPy array
eeg_signal = eeg_data.values

# Perform the wavelet decomposition for each channel
frequency_band_coefficients = {}

for channel in range(eeg_signal.shape[1]):
    # Extract the channel's EEG signal
    channel_signal = eeg_signal[:, channel]

    # Perform the wavelet decomposition with 'symmetric' mode
    coeffs = pywt.wavedec(channel_signal, wavelet, mode='symmetric')

    # Calculate the number of levels in the wavelet decomposition
    num_levels = len(coeffs) - 1

    # Calculate the frequency resolution at each level
    freq_resolutions = [128 / (2 ** (i + 1)) for i in range(num_levels)]
    print(freq_resolutions)
    # Extract the frequency bands from the detail coefficients
    channel_band_coefficients = {}

    for band, (start_freq, end_freq) in frequency_bands.items():
        print(band , start_freq , end_freq)
        # Initialize an empty list to store the band coefficients
        band_coefficients = []

        # Iterate through each level of the wavelet decomposition
        for level in range(num_levels):
            level_resolution = freq_resolutions[level]
            print(level_resolution)
            # Calculate the indices corresponding to the frequency band
            start_index = int(start_freq * level_resolution)
            end_index = int(end_freq *level_resolution)

            # Extract the coefficients within the frequency band
            level_coeffs = coeffs[level + 1]  # Skip the approximation coefficients
            band_coeffs = level_coeffs[start_index:end_index + 1]
            band_coefficients.extend(band_coeffs)

        channel_band_coefficients[band] = band_coefficients

    # Store the frequency band coefficients for the current channel
    frequency_band_coefficients[f'Channel {channel + 1}'] = channel_band_coefficients

```

```python

wavelet = 'db4'  # Example: Daubechies wavelet with 4 coefficients
level = 5        # Example: Decompose into 5 levels

# Calculate the number of samples
num_samples = eeg_data.shape[0]

# Set the wavelet type and decomposition level
# Perform wavelet decomposition to obtain approximation and detail coefficients
approximation_coeffs = []
alpha_coeffs = []
beta_coeffs = []
gamma_coeffs = []
theta_coeffs = []

for channel in channels:
    coeffs = pywt.wavedec(eeg_data[channel], wavelet, level=level)
    a_coeffs = coeffs[0]
    d_coeffs = coeffs[1:]

    approximation_coeffs.append(a_coeffs)
    alpha_coeffs.append(d_coeffs[0])
    beta_coeffs.append(d_coeffs[1])
    gamma_coeffs.append(d_coeffs[2])
    theta_coeffs.append(d_coeffs[3])

# Perform reconstruction to obtain the filtered signals
alpha_signals = []
beta_signals = []
gamma_signals = []
theta_signals = []

for channel in range(eeg_data.shape[1]):
    if len(alpha_coeffs[channel]) > 0:
        alpha_signal = pywt.waverec([alpha_coeffs[channel]], wavelet)
    else:
        alpha_signal = np.zeros(num_samples)

    if len(beta_coeffs[channel]) > 0:
        beta_signal = pywt.waverec([beta_coeffs[channel]], wavelet)
    else:
        beta_signal = np.zeros(num_samples)

    if len(gamma_coeffs[channel]) > 0:
        gamma_signal = pywt.waverec([gamma_coeffs[channel]], wavelet)
    else:
        gamma_signal = np.zeros(num_samples)

    if len(theta_coeffs[channel]) > 0:
        theta_signal = pywt.waverec([theta_coeffs[channel]], wavelet)
    else:
        theta_signal = np.zeros(num_samples)

    alpha_signals.append(alpha_signal)
    beta_signals.append(beta_signal)
    gamma_signals.append(gamma_signal)
    theta_signals.append(theta_signal)

# Optionally, you can normalize the signals to have zero mean
# alpha_signals = [signal - np.mean(signal) for signal in alpha_signals]
# beta_signals = [signal - np.mean(signal) for signal in beta_signals]
# gamma_signals = [signal - np.mean(signal) for signal in gamma_signals]
# theta_signals = [signal - np.mean(signal) for signal in theta_signals]
```
