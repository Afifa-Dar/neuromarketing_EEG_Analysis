```python
import numpy as np
import pywt
from scipy import signal

# Assuming you have EEG data stored in a numpy array called 'eeg_data'
# with a sampling frequency of 'fs' (in Hz)

# Choose the wavelet function for decomposition
wavelet_name = 'db4'

# Perform wavelet decomposition of EEG data
coeffs = pywt.wavedec(eeg_data, wavelet=wavelet_name, level=5)

# Extract the detail coefficients at each level
detail_coeffs = coeffs[1:]  # Exclude the approximation coefficients

# Print the detail coefficients at each level
for i, detail_coeff in enumerate(detail_coeffs):
    print(f"Detail coefficients at level {i+1}:")
    print(detail_coeff)
    print()




# Assuming you have detail coefficients stored in a list called 'detail_coeffs'

# Define the frequency ranges for each EEG band
delta_band = (0.5, 4)    # Delta band: 0.5 - 4 Hz
theta_band = (4, 8)      # Theta band: 4 - 8 Hz
alpha_band = (8, 13)     # Alpha band: 8 - 13 Hz
beta_band = (13, 30)     # Beta band: 13 - 30 Hz
gamma_band = (30, 100)   # Gamma band: 30 - 100 Hz

# Define the sampling frequency (fs) and filter order
fs = 256                # Sampling frequency in Hz
filter_order = 4        # Filter order

# Initialize empty arrays for filtered components
delta_filtered = []
theta_filtered = []
alpha_filtered = []
beta_filtered = []
gamma_filtered = []

# Apply bandpass filters to extract frequency bands from detail coefficients
for detail_coeff in detail_coeffs:
    # Apply bandpass filter for each EEG band
    delta_filtered.append(signal.sosfilt(signal.butter(filter_order, delta_band, fs=fs, btype='bandpass', output='sos'), detail_coeff))
    theta_filtered.append(signal.sosfilt(signal.butter(filter_order, theta_band, fs=fs, btype='bandpass', output='sos'), detail_coeff))
    alpha_filtered.append(signal.sosfilt(signal.butter(filter_order, alpha_band, fs=fs, btype='bandpass', output='sos'), detail_coeff))
    beta_filtered.append(signal.sosfilt(signal.butter(filter_order, beta_band, fs=fs, btype='bandpass', output='sos'), detail_coeff))
    gamma_filtered.append(signal.sosfilt(signal.butter(filter_order, gamma_band, fs=fs, btype='bandpass', output='sos'), detail_coeff))

# Print the shape of the filtered components for each frequency band
print("Delta filtered shape:", [component.shape for component in delta_filtered])
print("Theta filtered shape:", [component.shape for component in theta_filtered])
print("Alpha filtered shape:", [component.shape for component in alpha_filtered])
print("Beta filtered shape:", [component.shape for component in beta_filtered])
print("Gamma filtered shape:", [component.shape for component in gamma_filtered])
```
