```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
```

```python
models=[]
for i in range(1,6):
    models.append(tf.keras.models.load_model(f"models/modified-data-5Fold-5Model/fold_{i}"))
```

```python
chnl = [f'ch_{i}' for i in range(1,15) ]
rng = [-150,150]
```

```python
directory = '../like-dislike/clean_data/'

eeg_data = []
labels = []
for filename in os.scandir(directory):
    data = pd.read_csv(filename.path)
    for ch in chnl:
        data.loc[(data[ch] < rng[0]) | (data[ch] > rng[1]),ch] = np.nan
        data[ch].replace(
                to_replace=np.nan,
                value=data[ch].mean() ,inplace=True)
    if (data["label"][0] == 'Like'):
        labels.append(0)
    else:
        labels.append(1)
    eeg_data.append(data.drop(['label'] , axis = 1))
    
```

```python
eeg_data = np.array(eeg_data)
eeg_data.shape
```

```python
labels = np.array(labels)
labels.shape
```

```python
for i in models:
    i.fit(eeg_data , labels , epochs=50 , batch_size = 16)
```

```python
y_pred = []
for data in eeg_data:
    res= []
    for modl in models:
        res.append(np.argmax(modl.predict(np.array([data,]))))
    print(res)
    
    if res.count(0) > res.count(1):
        y_pred.append(0)
    else:
        y_pred.append(1)
        
```

```python
print(y_pred.count(0) , y_pred.count(1))
```

```python
import sklearn 
```

```python
len(labels)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix 
```

```python
cm = confusion_matrix(labels,y_pred)
 
#Plot the confusion matrix.
sns.heatmap(cm,
            annot=True,
            fmt='g',
            xticklabels=['Like','Dislike'],
            yticklabels=['Like','Dislike'])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python
accuaracy = (cm[0][0] + cm[1][1]) / eeg_data.shape[0] *100
recall = (cm[0][0]) /(cm[0][0] + cm[1][0]) *100
precision = (cm[0][0]) /(cm[0][0] + cm[0][1]) *100
f1_score = (2*recall*precision)/(recall + precision)
```

```python
print('Accuracy: ' , accuaracy)
print('Precision: ' , precision)
print('Recall: ' , recall)
print('F1 SCore: ' , f1_score)
```
