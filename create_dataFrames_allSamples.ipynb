```python
import pandas as pd
import os
```

```python
labels = []
```

```python
 
# assign directory
directory = 'labels'
 
# iterate over files in
# that directory
for filename in os.scandir(directory):
    lab = open(filename.path , "r").read()
    if lab == 'Disike':
        lab = 'Dislike'
    labels.append(lab)
```

```python
len(labels)
```

```python
l = 0
d = 0
for i in labels:
    if i == 'Like':
        l+=1
    else:
        d+=1
print('like',l,'dislike',d)
```

```python
# import required module
import os
 
# assign directory
directory = '25-users'
l = -1
c = 1
# iterate over files in
# that directory
for filename in os.listdir(directory):
    l+=1
    data = pd.read_csv(f'{directory}/{filename}', sep=" ", header=None, names=channels)
    data['label'] = labels[l]
    data.to_csv(f'datasets/{filename[:-4]}.csv',index = False)
    c+=1
   
```
