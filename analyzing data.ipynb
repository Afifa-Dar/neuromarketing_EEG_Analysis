```python
import pandas as pd
import numpy as np
import os
```

```python
channels = [f'ch_{i}' for i in range(1,15) ]

```

```python
dislike_range = {
                "ch_1" : [-150,150] ,
                 'ch_2':[-100 , 100] , 
                 'ch_3':[-100 , 100],
                 'ch_4':[-100 , 100],
                 'ch_5':[-20,20],
                 'ch_6':[-120 , 120] ,
                 'ch_7':[-150 , 150] ,
                 'ch_8': [-140 ,140 ] , 
                 'ch_9':[-150, 150] , 
                 'ch_10':[-20 ,20] , 
                 'ch_11':[-130 ,130] , 
                 'ch_12':[-100 , 100] , 
                 'ch_13':[-100 , 100] , 
                 'ch_14':[-100 , 100]
}
like_range = {
                "ch_1" : [-60,60] ,
                 'ch_2':[-60 , 60] , 
                 'ch_3':[-100 , 100],
                 'ch_4':[-50 , 50],
                 'ch_5':[-10,10],
                 'ch_6':[-50 , 50] ,
                 'ch_7':[-50 , 50] ,
                 'ch_8': [-40 ,40 ] , 
                 'ch_9':[-50, 50] , 
                 'ch_10':[-10 ,10] , 
                 'ch_11':[-50 ,50] , 
                 'ch_12':[-50 , 50] , 
                 'ch_13':[-50 , 50] , 
                 'ch_14':[-40 , 40]
}
```

```python
directory = 'clean_data/'

faulted_data = {}
count = 0
for i in os.listdir(directory):
    data = pd.read_csv(f'{directory}{i}')
    label = data['label'][0]
    faulted_data[i] = {}
    for ch in channels:
        faulted_data[i][ch] = 0
        d = data[ch]  
        if label == 'Like':
            rng = like_range[ch]
        else:
            rng = dislike_range[ch]
        
        for dv in d:
            if dv < rng[0] or dv > rng[1]:
                faulted_data[i][ch]+=1
        if faulted_data[i][ch] == 0:
            del faulted_data[i][ch]
    if faulted_data[i] == {}:
            del faulted_data[i]
```
